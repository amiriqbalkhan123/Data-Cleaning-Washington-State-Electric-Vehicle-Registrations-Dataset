```python
# Function to remove all the leading / trailing whitespaces from the columns:

def remove_spaces(df, columns=None):
    df = df.copy()
    if columns is None:
        columns = df.select_dtypes(include=['object', 'string']).columns.tolist()
    for col in columns:
        if col in df.columns and pd.api.types.is_string_dtype(df[col]):
            df[col] = df[col].str.strip()
    return df

cols_to_remove_spaces = ['County','City', 'State', 'Postal Code', 'Model Year', 
                        'Make', 'Electric Vehicle Type', 'cafv_eligibility',
                        'Electric Range', 'Base MSRP','Legislative District',
                        'DOL Vehicle ID','Electric Utility','2020 Census Tract',
                        'Vehicle Longtitude','Vehicle Latitude']

df = remove_spaces(df, cols_to_remove_spaces)
```
